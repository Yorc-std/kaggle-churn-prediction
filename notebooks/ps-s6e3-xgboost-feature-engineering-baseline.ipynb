This notebook presents a clean **weighted blend version** for the **Playground Series S6E3 Customer Churn** competition.  
The goal is simple: combine multiple strong submission files into a single, more stable final prediction.

Unlike the previous notebook, **PS S6E3 | XGBoost + Feature Engineering Baseline**, this version does **not train a new model**.  
Instead, it focuses on **submission ensembling**, where carefully selected weights are used to improve leaderboard performance.

```python
from pathlib import Path
import pandas as pd

def blend_submissions(weight_dict, output_path, target_col="Churn"):
    blended = None
    total_weight = sum(weight_dict.values())

    for path, weight in weight_dict.items():
        df = pd.read_csv(path)[["id", target_col]].copy()
        df[target_col] *= weight

        if blended is None:
            blended = df
        else:
            blended[target_col] += df[target_col]

    blended[target_col] /= total_weight
    blended.to_csv(output_path, index=False)
    print(f"Saved to {output_path}")

BASE_PATH = Path("/kaggle/input/datasets/anthonytherrien/predict-customer-churn-vault")

weight_dict = {
    str(BASE_PATH / "submission.csv"): 0.95,
    str(BASE_PATH / "submission (1).csv"): 0.05,
}

blend_submissions(weight_dict, "submission.csv")
```
